# The _ _init_ _() Function


To understand the meaning of classes we have to understand the built-in _ _init_ _() function which is a special function.

All classes have a function called _ _init_ _(), which is always executed when the class is being initiated.

Use the _ _init_ _() function to assign values to object properties, or other operations that are necessary to do when the object is being created:

```python
# Defining a simple class "Person"
class Person:
    # Constructor to initialize Attributes
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # Define Methods
    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")

# Creating objects (instances) of the class "Person"
person1 = Person("Alice", 30)
person2 = Person("Bob", 25)

# Accessing attributes and methods of the objects
print(person1.name)  # Output: "Alice"
print(person2.age)   # Output: 25
person1.greet()      # Output: "Hello, my name is Alice and I am 30 years old."
person2.greet()      # Output: "Hello, my name is Bob and I am 25 years old."
```

###### Note: The __init__() function is called automatically every time the class is being used to create a new object.

self is coming automatically ("no need to type")

init compulsory has self as argument.

init method is used to initialize variables.

###### Note: The self parameter is a reference to the current instance of the class, and is used to access variables that belong to the class.

## Types of constructors : 

Default constructor: The default constructor is a simple constructor which doesnâ€™t accept any arguments. Its definition has only one argument which is a reference to the instance being constructed.

Parameterized constructor: constructor with parameters is known as parameterized constructor. The parameterized constructor takes its first argument as a reference to the instance being constructed known as self and the rest of the arguments are provided by the programmer.

```python
class students:
 
    # default constructor
    def __init__(self):
        self.geek = "ECE students"
    
    # a method for printing data members
    def print_ece(self):
        print(self.geek)
        
# creating object of the class
obj = students()
 
# calling the instance method using the object obj
obj.print_ece()
```

```python
class Addition:
    first = 0
    second = 0
    answer = 0
     
    # parameterized constructor
    def __init__(self, f, s):
        self.first = f
        self.second = s
     
    def display(self):
        print("First number = " + str(self.first))
        print("Second number = " + str(self.second))
        print("Addition of two numbers = " + str(self.answer))
 
    def calculate(self):
        self.answer = self.first + self.second
 
# creating object of the class
# this will invoke parameterized constructor
obj = Addition(1000, 2000)
 
# perform Addition
obj.calculate()
 
# display result
obj.display()
```

### Adding and Retrieving attributes Dynamically

```python
class MyClass:
    def __init__(self, name):
        self.name = name

# Create an instance
obj = MyClass("Example")

# Directly add an attribute
obj.age = 25

# Retrieve the added attribute
print(obj.age)  # Output: 25
```

```python
class MyClass:
    def __init__(self, name):
        self.name = name

# Create an instance
obj = MyClass("Example")

# Dynamically add an attribute
setattr(obj, 'age', 25)

# Retrieve the dynamically added attribute
print(getattr(obj, 'age'))  # Output: 25
```

```python
class MyClass:
    def __init__(self, name):
        self.name = name

    def add_attribute(self, key, value):
        setattr(self, key, value)

# Create an instance
obj = MyClass("Example")

# Use the method to add attributes dynamically
obj.add_attribute('age', 25)
obj.add_attribute('location', "New York")

# Retrieve the dynamic attributes
print(obj.age)        # Output: 25
print(obj.location)   # Output: New York
```

```python
class MyClass:
    def __init__(self, name):
        self.name = name

# Create an instance
obj = MyClass("Example")

# Add dynamic attributes using __dict__
obj.__dict__['age'] = 25
obj.__dict__['location'] = "New York"

# Retrieve the dynamic attributes
print(obj.__dict__['age'])        # Output: 25
print(obj.__dict__['location'])   # Output: New York
```

### Python Inheritance: 
Inheritance allows us to create a new class from an existing class.

The new class that is created is known as subclass (child or derived class) and the existing class from which the child class is derived is known as superclass (parent or base class).

```python
class Animal:
    # attribute and method of the parent class
    name = ""
    def eat(self):
        print("I can eat")

# inherit from Animal
class Dog(Animal):
    # new method in subclass
    def display(self):
        # access name attribute of superclass using self
        print("My name is ", self.name)

# create an object of the subclass
la = Dog()
# access superclass attribute and method 
la.name = "Rohu"
la.eat()
# call subclass method 
la.display()
```

```python
## Example of Inheritance 

class Polygon:
    def __init__(self, no_of_sides):
        self.n = no_of_sides
        self.sides = [0 for i in range(no_of_sides)]

    def inputSides(self):
        self.sides = [float(input("Enter side "+str(i+1)+" : ")) for i in range(self.n)]

    def dispSides(self):
        for i in range(self.n):
            print("Side",i+1,"is",self.sides[i])
            
poly = Polygon(4)    
poly.inputSides()
poly.dispSides()
```

A triangle is a polygon with 3 sides. So, we can create a class called Triangle which inherits from Polygon. This makes all the attributes of Polygon class available to the Triangle class.

We don't need to define them again (code reusability). Triangle can be defined as follows.

```python
class Triangle(Polygon):
    def __init__(self):
        Polygon.__init__(self,3)

    def findArea(self):
        a, b, c = self.sides
        # calculate the perimeter
        s = (a + b + c) 
        print('The perimeter of the triangle is %0.2f' %s)
        area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
        print('The area of the triangle is %0.2f' %area)
        
# Creating an instance of the Triangle class
t = Triangle()

# Prompting the user to enter the sides of the triangle
t.inputSides()

# Displaying the sides of the triangle
t.dispSides()

# Calculating and printing the area of the triangle
t.findArea()
```

```python
class Animal:
    # attributes and method of the parent class
    def eat(self):
        print("I can eat")

# inherit from Animal
class Dog(Animal):
    # override eat() method
    def eat(self):
        print("I like to eat bones")

# create an object of the subclass
l = Dog()
# call the eat() method on the l object
l.eat()
```

```python
class Animal:
    def eat(self):
        print("I can eat")

# inherit from Animal
class Dog(Animal):
    # override eat() method
    def eat(self):
        # call the eat() method of the superclass using super()
        super().eat()
        print("I like to eat bones")

# create an object of the subclass
la = Dog()
la.eat()
```

## Polymorphism:

There are different methods to use polymorphism in Python. You can use different functions, class methods or objects to define polymorphism. 

```python
class Tomato(): 
     def type(self): 
       print("Vegetable") 
     def color(self):
       print("Red") 
class Apple(): 
     def type(self): 
       print("Fruit") 
     def color(self): 
       print("Yellow") 
      
def func(obj): 
    obj.type() 
    obj.color()
        
obj_tomato = Tomato() 
obj_apple = Apple() 
func(obj_tomato) 
func(obj_apple)
```

```python
class India():
     def capital(self):
       print("New Delhi")
 
     def language(self):
       print("Hindi and English")
 
class USA():
     def capital(self):
       print("Washington, D.C.")
 
     def language(self):
       print("English")
 
obj_ind = India()
obj_usa = USA()
for country in (obj_ind, obj_usa):
    country.capital()
    country.language()
```

```python
class Bird:
     def intro(self):
       print("There are different types of birds")
 
     def flight(self):
       print("Most of the birds can fly but some cannot")
 
class parrot(Bird):
     def flight(self):
       print("Parrots can fly")
 
class penguin(Bird):
     def flight(self):
       print("Penguins do not fly")
 
obj_bird = Bird()
obj_parr = parrot()
obj_peng = penguin()
 
obj_bird.intro()
obj_bird.flight()
 
obj_parr.intro()
obj_parr.flight()
 
obj_peng.intro()
obj_peng.flight()
```

### Overloading and Overlapping Operators:

```python
#Eg:1 Overloading the __add__ Method
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

v1 = Vector(1, 2)
v2 = Vector(3, 4)
result = v1 + v2
print(result.x, result.y)  # Output: 4 6
```

```python
#Eg:2 Overloading the __mul__ Method (Multiplication):
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __mul__(self, scalar):
        return Vector(self.x * scalar, self.y * scalar)

v = Vector(2, 3)
result = v * 3
print(result.x, result.y)  # Output: 6 9
```

### Data modeling examples 

```python
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

class Customer:
    def __init__(self, name, email, address):
        self.name = name
        self.email = email
        self.address = address

class Order:
    def __init__(self, order_id, customer, products):
        self.order_id = order_id
        self.customer = customer
        self.products = products
        self.total_price = sum(product.price for product in products)

    def __str__(self):
        product_list = "\n".join(f"{product.name}: ${product.price}" for product in self.products)
        return f"Order ID: {self.order_id}\nCustomer: {self.customer.name}\nProducts:\n{product_list}\nTotal Price: ${self.total_price}"

# Sample data
product1 = Product("Keyboard", 50, 3)
product2 = Product("Mouse", 20, 5)
customer = Customer("Alice", "alice@example.com", "123 Main St")
order = Order(1, customer, [product1, product2])

print(order)
# The __str__ method is intended to provide a user-friendly
# string representation of the object. 
# This method is called when you use print() or str() 
# on an instance of the class.
```

```python
class Course:
    def __init__(self, title, instructor, price):
        self.title = title
        self.instructor = instructor
        self.price = price

    def __str__(self):
        return f"Course: {self.title}\nInstructor: {self.instructor}\nPrice: ${self.price}"

class Student:
    def __init__(self, name, email):
        self.name = name
        self.email = email

class Enrollment:
    def __init__(self, student, course):
        self.student = student
        self.course = course

    def __str__(self):
        return f"Enrollment: {self.student.name} in {self.course.title}"

# Sample data
course1 = Course("Python Programming", "John Doe", 50)
course2 = Course("Web Development", "Jane Smith", 80)
student1 = Student("Alice", "alice@example.com")
student2 = Student("Bob", "bob@example.com")
enrollment1 = Enrollment(student1, course1)
enrollment2 = Enrollment(student2, course2)

print(course1)
print(enrollment1)
```

```python
class LibraryItem:
    def __init__(self, title, item_id):
        self.title = title
        self.item_id = item_id

    def get_details(self):
        return f"Title: {self.title}, ID: {self.item_id}"

# Derived class for Books
class Book(LibraryItem):
    def __init__(self, title, item_id, author, pages):
        super().__init__(title, item_id)
        self.author = author
        self.pages = pages

    def get_details(self):
        base_details = super().get_details()
        return f"{base_details}, Author: {self.author}, Pages: {self.pages}"

# Derived class for DVDs
class DVD(LibraryItem):
    def __init__(self, title, item_id, director, duration):
        super().__init__(title, item_id)
        self.director = director
        self.duration = duration

    def get_details(self):
        base_details = super().get_details()
        return f"{base_details}, Director: {self.director}, Duration: {self.duration} minutes"

book = Book("1984", 1, "George Orwell", 328)
dvd = DVD("Inception", 2, "Christopher Nolan", 148)

print(book.get_details())  # Output: Title: 1984, ID: 1, Author: George Orwell, Pages: 328
print(dvd.get_details())  # Output: Title: Inception, ID: 2, Director: Christopher Nolan, Duration: 148 minutes
```

```python
class Employee:
    def __init__(self, name, employee_id, department):
        self.name = name
        self.employee_id = employee_id
        self.department = department

    def get_details(self):
        return f"Name: {self.name}, ID: {self.employee_id}, Department: {self.department}"

# Derived class for Full-time Employees
class FullTimeEmployee(Employee):
    def __init__(self, name, employee_id, department, salary):
        super().__init__(name, employee_id, department)
        self.salary = salary

    def get_details(self):
        base_details = super().get_details()
        return f"{base_details}, Salary: {self.salary}"

# Derived class for Part-time Employees
class PartTimeEmployee(Employee):
    def __init__(self, name, employee_id, department, hourly_rate):
        super().__init__(name, employee_id, department)
        self.hourly_rate = hourly_rate

    def get_details(self):
        base_details = super().get_details()
        return f"{base_details}, Hourly Rate: {self.hourly_rate}"

ft_employee = FullTimeEmployee("Alice", 123, "Engineering", 70000)
pt_employee = PartTimeEmployee("Bob", 456, "Marketing", 30)

print(ft_employee.get_details())  # Output: Name: Alice, ID: 123, Department: Engineering, Salary: 70000
print(pt_employee.get_details())  # Output: Name: Bob, ID: 456, Department: Marketing, Hourly Rate: 30
```
