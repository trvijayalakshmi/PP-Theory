### Python Inheritance: 

```python
Types of Inheritance:
    1. Single inheritance
    2. Multiple Inheritance
    3. Multilevel Inheritance
    4. Hierarchical Inheritance
```

```python
class Animal:
    def speak(self):
        print("This is an animal.")

class Dog(Animal):   # Single Inheritance
    def bark(self):
        print("Dog barks.")

d = Dog()   #Object creation
d.speak()   # use of parent method
d.bark()    # use of child method
```

```python
class A:      # Parent 1
    def method_A(self):
        print("Method from A")

class B:     # Parent 2
    def method_B(self):
        print("Method from B")

class C(A, B):   # Multiple Inheritance
    def add(self, a, b):        
        print("sum of a and b is ", a+b)

obj_c = C()    # Object creation from child class
obj_c.method_A()   # accessing method of parent class A
obj_c.method_B()   # accessing method of parent class B
obj_c.add(4,6)      # accessing method of Child class C
```

```python
class Grandparent:
    def show_gp(self):
        print("I am Grandparent")

class Parent(Grandparent):
    def show_p(self):
        print("I am Parent")

class Child(Parent):
    def show_c(self):
        print("I am Child")

c = Child()
c.show_gp()
c.show_p()
c.show_c()
```

```python
class Parent:
    def greet(self):
        print("Hello from Parent")

class Child1(Parent):
    def peri(self, a, b):
        d= 2*(a+b)
        print("Result of Child1 method ", d)

class Child2(Parent):
    def area(self, h, b):
        a = h * b
        print("Result of Child2 method ", a)

c1 = Child1()
c2 = Child2()
c1.greet()
c2.greet()
c1.peri(4,3)
c2.area(4,3)
```

```python
class Polygon:
    def __init__(self, no_of_sides):
        self.n = no_of_sides
        self.sides = [0 for i in range(no_of_sides)]

    def inputSides(self):
        self.sides = [float(input("Enter side "+str(i+1)+" : ")) for i in range(self.n)]

    def dispSides(self):
        for i in range(self.n):
            print("Side",i+1,"is",self.sides[i])
            
poly = Polygon(4)    
poly.inputSides()
poly.dispSides()
```

```python
class Triangle(Polygon):
    def __init__(self):
        Polygon.__init__(self,3)

    def findArea(self):
        a, b, c = self.sides
        # calculate the perimeter
        s = (a + b + c) 
        print('The perimeter of the triangle is %0.2f' %s)
        area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
        print('The area of the triangle is %0.2f' %area)
        
# Creating an instance of the Triangle class
t = Triangle()

# Prompting the user to enter the sides of the triangle
t.inputSides()

# Displaying the sides of the triangle
t.dispSides()

# Calculating and printing the area of the triangle
t.findArea()
```

```python
Ex1: Overriding method
```

```python
class Parent:    # Superclass
    def display(self):
        print("Display from Parent")

class Child(Parent):   # Subclass
    def display(self):    # Overriding
        print("Display from Child")

c = Child()  # Object creation
c.display()  
```

```python
class Parent:    # Superclass
    def display(self):
        print("Display from Parent")

class Child(Parent):   # Subclass
    def display(self):    # Overriding
#         super().display()   
        Parent.display(self)
        print("Display from Child")

c = Child()  # Object creation
c.display()  
```

```python
Ex2: Overriding method
```

```python
class Animal:   #Superclass
    # attributes and method of the parent class
    def eat(self):
        print("I can eat")

# inherit from Animal 
class Dog(Animal):   #Subclass
    def eat(self):
        print("I like to eat bones")

# create an object of the Subclass
s = Dog()
s.eat()
```

```python
class Animal:
    def eat(self):
        print("I can eat")

# inherit from Animal
class Dog(Animal):
    # override eat() method
    def eat(self):
#         super().eat()
        Animal.eat(self)
        print("I like to eat bones")

# create an object of the subclass
la = Dog()
la.eat()
```

```python
class Calculator:
    def add(self, a=0, b=0, c=0):
        return a + b + c

c = Calculator()
print(c.add(2, 3))       # 5
print(c.add(2, 3, 4))    # 9
print(c.add(5))          # 5
```

```python
class Display:
    def show(self, value):
        if isinstance(value, int):
            print("Integer:", value)
        elif isinstance(value, str):
            print("String:", value)
        else:
            print("Unknown type:", value)

d = Display()
d.show(10)          # Integer: 10
d.show("Python")    # String: Python
d.show(10.7)
d.show([1, 2, 3])   # Unknown type: [1, 2, 3]
```
