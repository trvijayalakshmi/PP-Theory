### Python Inheritance: 

```python
Types of Inheritance:
    1. Single inheritance
    2. Multiple Inheritance
    3. Multilevel Inheritance
    4. Hierarchical Inheritance
```

```python
class Animal:
    def speak(self):
        print("This is an animal.")

class Dog(Animal):   # Single Inheritance
    def bark(self):
        print("Dog barks.")

d = Dog()   #Object creation
d.speak()   # use of parent method
d.bark()    # use of child method
```

```python
class A:      # Parent 1
    def method_A(self):
        print("Method from A")

class B:     # Parent 2
    def method_B(self):
        print("Method from B")

class C(A, B):   # Multiple Inheritance
    def add(self, a, b):        
        print("sum of a and b is ", a+b)

obj_c = C()    # Object creation from child class
obj_c.method_A()   # accessing method of parent class A
obj_c.method_B()   # accessing method of parent class B
obj_c.add(4,6)      # accessing method of Child class C
```

```python
# Base Class 1
class Teacher:
    def __init__(self, name):
        self.name = name

    def teach(self):
        print(f"{self.name} is teaching.")

# Base Class 2
class Researcher:
    def __init__(self, field):
        self.field = field

    def research(self):
        print(f"Researching in the field of {self.field}.")

# Derived Class (inherits from both Teacher and Researcher)
class Professor(Teacher, Researcher):
    def __init__(self, name, field):
        Teacher.__init__(self, name)      # Initialize Teacher part
        Researcher.__init__(self, field)  # Initialize Researcher part

    def display(self):
        print(f"Professor {self.name}, Field: {self.field}")

# Object creation
p = Professor("Dr. Anita", "Artificial Intelligence")
p.display()
p.teach()
p.research()
```

```python
class Grandparent:
    def show_gp(self):
        print("I am Grandparent")

class Parent(Grandparent):
    def show_p(self):
        print("I am Parent")

class Child(Parent):
    def show_c(self):
        print("I am Child")

c = Child()
c.show_gp()
c.show_p()
c.show_c()
```

```python
class Person:
    def __init__(self, name): self.name = name

class Employee(Person):  # Employee (inherits from Person)
    def __init__(self, name, emp_id):
        super().__init__(name)
        self.emp_id = emp_id  # additional attribute: emp_id

class Manager(Employee):   # Manager (inherits from Employee)
    def __init__(self, name, emp_id, dept):
        super().__init__(name, emp_id)
        self.dept = dept  # additional attribute: dept.

    def display(self):
        print(f"Name: {self.name}, ID: {self.emp_id}, Dept: {self.dept}")

m = Manager("Rajesh", 101, "ECE")
m.display()
```

```python
class Parent:
    def greet(self):
        print("Hello from Parent")

class Child1(Parent):
    def peri(self, a, b):
        d= 2*(a+b)
        print("Result of Child1 method ", d)

class Child2(Parent):
    def area(self, h, b):
        a = h * b
        print("Result of Child2 method ", a)

c1 = Child1()
c2 = Child2()
c1.greet()
c2.greet()
c1.peri(4,3)
c2.area(4,3)
```

```python
class Polygon:
    def __init__(self, no_of_sides):
        self.n = no_of_sides
        self.sides = [0 for i in range(no_of_sides)]

    def inputSides(self):
        self.sides = [float(input("Enter side "+str(i+1)+" : ")) for i in range(self.n)]

    def dispSides(self):
        for i in range(self.n):
            print("Side",i+1,"is",self.sides[i])
            
poly = Polygon(4)    
poly.inputSides()
poly.dispSides()
```

```python
class Triangle(Polygon):
    def __init__(self):
        Polygon.__init__(self,3)

    def findArea(self):
        a, b, c = self.sides
        # calculate the perimeter
        s = (a + b + c) 
        print('The perimeter of the triangle is %0.2f' %s)
        area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
        print('The area of the triangle is %0.2f' %area)
        
# Creating an instance of the Triangle class
t = Triangle()

# Prompting the user to enter the sides of the triangle
t.inputSides()

# Displaying the sides of the triangle
t.dispSides()

# Calculating and printing the area of the triangle
t.findArea()
```
