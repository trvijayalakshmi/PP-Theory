### 3.  Lambda Functions: 

```python
Anonymous (nameless) one-line functions using the lambda keyword.
```

```python
Syntax: lambda arguments: expression
```

```python
square = lambda n: n * n
print(square(5)) 

add = lambda a, b: a + b
print(add(3, 7)) 
```

```python
numbers = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x**2, numbers))
print(squares) 
```

```python
nums = [10, 15, 20, 25, 30]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)
```

```python
from functools import reduce

nums = [1, 2, 3, 4, 5]
product = reduce(lambda a, b: a * b, nums)
print(product)  # 120 (1*2*3*4*5)
```

```python
from functools import reduce

nums = [1, 2, 3, 4, 5]

# Map → Squares
squares = list(map(lambda x: x**2, nums))
print("Squares:", squares)

# Filter → Even numbers
evens = list(filter(lambda x: x % 2 == 0, nums))
print("Evens:", evens)

# Reduce → Product
product = reduce(lambda a, b: a * b, nums)
print("Product:", product)
```

### 4. Programs on Lambda Functions

```python
maximum = lambda a, b: a if a > b else b
print(maximum(12, 8))  
```

```python
check_even = lambda x: "Even" if x % 2 == 0 else "Odd"
print(check_even(7))   
```

```python
pairs = [(1, 5), (3, 1), (6, 8), (2, 4)]
sorted_pairs = sorted(pairs, key=lambda x: x[1])
print(sorted_pairs)  
```

```python
names = ["Raj", "Ananya", "Kiran", "Jo"]
sorted_names = sorted(names, key=lambda x: len(x))
print(sorted_names)  # Expected output: ['Jo', 'Raj', 'Kiran', 'Ananya']
```

```python
words = ["Apple", "Banana", "Avocado", "Mango", "Apricot"]
a_words = list(filter(lambda w: w.startswith("A"), words))
print(a_words)
```

```python
nums = [2, 4, 6, 8]
tripled = list(map(lambda x: x*3, nums))
print(tripled)   # Expected output: [6, 12, 18, 24]
```

```python
nums = [-5, 3, -1, 0, 7, -9]
positive = list(filter(lambda x: x > 0, nums))
print(positive)   # Expected output: [3, 7]
```

```python
from functools import reduce

n = 5
factorial = reduce(lambda x, y: x * y, range(1, n+1))
print("Factorial of", n, "=", factorial)
```
