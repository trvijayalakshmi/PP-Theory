### 3.  Lambda Functions: 

```python
Anonymous (nameless) one-line functions using the lambda keyword.
```

```python
Lambda Syntax: lambda arguments: expression
```

```python
square = lambda n: n * n
print(square(5)) 

add = lambda a, b: a + b
print(add(3, 7)) 
```

```python
maximum = lambda a, b: a if a > b else b
print(maximum(12, 8))  
```

```python
check_even = lambda x: "Even" if x % 2 == 0 else "Odd"
print(check_even(7))   
```

```python
pairs = [(1, 5), (3, 1), (6, 8), (2, 4)]
sorted_pairs = sorted(pairs, key=lambda x: x[1])
print(sorted_pairs)  

sor = sorted(pairs)
print(sor)
```

```python
names = ["Raj", "Ananya", "Kiran", "Jo"]
sorted_names = sorted(names, key=lambda x: len(x))
print(sorted_names)  

sor= sorted(names)
print(sor)
```

### a. Map Function: Applies a function to each element of an iterable.

```python
Map Syntax:  map(lambda arguments: expression, iterable)
```

```python
numbers = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x**2, numbers))
print(squares) 
```

```python
Without lambda 
```

```python
def square(x):
    return x * x

print(list(map(square, [1, 2, 3, 4, 5])))
```

```python
nums = [10, 20, 30]
result = list(map(lambda x: x + 10, nums))
print(result)

result2 = list(map(lambda x: x + 10, range(1,5)))
print(result2)
```

```python
a = [1, 2, 3]
b = [4, 5, 6]
result = list(map(lambda x, y: x + y, a, b))
print(result)
```

```python
nums = [2, 4, 6, 8]
tripled = list(map(lambda x: x*3, nums))
print(tripled)   

t = list(map(lambda x: x*3, range(1,5)))
print(t)
```

### b. Filter Function: Filters elements based on a condition (True/False).

```python
Filter Syntax: filter(lambda arguments: condition, iterable)
```

```python
nums = [10, 15, 20, 25, 30]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)
```

```python
words = ["Apple", "Banana", "Avocado", "Mango", "Apricot"]
a_words = list(filter(lambda w: w.startswith("A"), words))
print(a_words)
```

```python
nums = [-5, 3, -1, 0, 7, -9]
positive = list(filter(lambda x: x > 0, nums))
print(positive)   # Expected output: [3, 7]
```

### c. Reduce Function: Reduces an iterable to a single value.

Requires functools.reduce.

```python
Reduce Syntax: reduce(lambda arguments: expression, iterable)
```

```python
from functools import reduce

nums = [1, 2, 3, 4, 5]
product = reduce(lambda a, b: a + b, nums)
print(product)  
```

```python
from functools import reduce

nums = [1, 2, 3, 4, 5]
product = reduce(lambda a, b: a * b, nums)
print(product) 
```

```python
from functools import reduce

n = 5
factorial = reduce(lambda x, y: x * y, range(1, n+1))
print("Factorial of", n, "=", factorial)
```

```python
from functools import reduce

numbers = [10, 45, 2, 99, 23]
maximum = reduce(lambda x, y: x if x > y else y, numbers)
print(maximum)
```

```python
from functools import reduce

nums = [1, 2, 3, 4, 5]

# Map → Squares
squares = list(map(lambda x: x**2, nums))
print("Squares:", squares)

# Filter → Even numbers
evens = list(filter(lambda x: x % 2 == 0, nums))
print("Evens:", evens)

# Reduce → Product
product = reduce(lambda a, b: a * b, nums)
print("Product:", product)
```

```python
filter()	Select elements based on condition
map()	Modify each element
reduce()	Combine all elements into one value
```
