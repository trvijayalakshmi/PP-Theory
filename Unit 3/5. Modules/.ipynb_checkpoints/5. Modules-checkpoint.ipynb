### 1. Importing Built-in Modules

```python
import math

print("Square root of 16 is:", math.sqrt(16))
print("Value of pi:", math.pi)
```

```python
import math as mh

print("Square root of 16 is:", mh.sqrt(16))
print("Value of pi:", mh.pi)
```

### 2. Creating a User-defined Module

```python
def add(a, b):
    return a + b

def sub(a,b):
    return a - b

def mul(a,b):
    return a * b

def div(a,b):
    return a / b
```

```python
import arithmetic

print("Sum:", arithmetic.add(10, 20))
print("Diff:", arithmetic.sub(20,40))
print("Product:", arithmetic.mul(20,40))
print("Division:", arithmetic.div(40,20))
```

```python
import arithmetic as mym

print("Sum:", mym.add(10, 20))
print("Diff:", mym.sub(20,40))
print("Product:", mym.mul(20,40))
print("Division:", mym.div(40,20))
```

```python
from arithmetic import add, mul

print("Sum:", add(10, 20))
print("Diff:", sub(20,40))
print("Product:", mul(20,40))
print("Division:", div(40,20))
```

```python
from arithmetic import *

print("Sum:", add(10, 20))
print("Diff:", sub(20,40))
print("Product:", mul(20,40))
print("Division:", div(40,20))
```

### 3. Standard Modules in Python

```python
import math

print("Square root of 25:", math.sqrt(25))
print("Factorial of 5:", math.factorial(5))
print("Ceil of 4.2:", math.ceil(4.2))
print("Floor of 4.8:", math.floor(4.8))
print("Value of pi:", math.pi)
print("Power (2^5):", math.pow(2, 5))
```

```python
import random

print("Random number between 1 and 10:", random.randint(1, 10))
print("Random float between 0 and 1:", random.random())
print("Random choice from list:", random.choice(['apple', 'banana', 'cherry']))
print("Shuffled list:", end=" ")
numbers = [1, 2, 3, 4, 5]
random.shuffle(numbers)
print(numbers)
```

```python
import datetime

# Current date and time
now = datetime.datetime.now()
print("Current Date and Time:", now)

# Specific date
d = datetime.date(2025, 8, 21)
print("Custom Date:", d)

# Extract components
print("Year:", now.year)
print("Month:", now.month)
print("Day:", now.day)
```
