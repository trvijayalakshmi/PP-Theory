### 5. Programs on Functions

```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

# Test the function
result = gcd(68,4)
print("GCD is ", result)
print("GCD of 64 and 4 is:", gcd(64, 4))   # Output: 4
print("GCD of 25 and 15 is:", gcd(25, 15)) # Output: 5
```

```python
def fibonacci_sequence(n):
    a, b = 0, 1
    count = 0
    while count < n:
        print(a, end=" ")
        a, b = b, a + b
        count += 1

# Driver code
terms = int(input("Enter the number of terms: "))
print("Fibonacci sequence:")
fibonacci_sequence(terms)
```

```python
def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact

print("Factorial of 5 is:", factorial(5))
```

```python
# function which return reverse of a string
def isPalindrome(s):
    return s == s[::-1]

s = "malayalam"
print(isPalindrome(s))
```

```python
def isPalindrome(st):
    for i in range(0, int(len(st)//2)):
        if st[i] != st[len(st)-i-1]:
            return False
    return True
 
s = "malayalam"
print(isPalindrome(s))
if isPalindrome(s):
    print("The string is a palindrome")
else:
    print("The string is not a palindrome")
```

```python
def prime(x, y):
    for i in range(x, y):
        if i == 0 or i == 1:
            continue
        else:
            for j in range(2, int(i/2)+1):
                if i % j == 0:
                    break
            else:
                print(i,end=" ")

# Driver program
starting_range = 6
ending_range = 30
prime(starting_range, ending_range)
```

```python
def swap(x, y): 
    temp = x; 
    x = y; 
    y = temp; 
  
# Driver code 
x = 2
y = 3
swap(x, y) 
print(x) 
print(y) 
```

```python
def harmonic_sum(n):
    total = 0.0
    for i in range(1, n + 1):
        total += 1 / i
    return total

# Example usage
n = int(input("Enter a number: "))
print(f"Harmonic sum of {n} = {harmonic_sum(n)}")
```

```python
# one line using sum() and list comprehension:
n = int(input("Enter a number: "))
harmonic = sum(1/i for i in range(1, n+1))
print(harmonic)
```
