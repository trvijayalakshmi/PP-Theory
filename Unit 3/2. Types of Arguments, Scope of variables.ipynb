### 1. Types of Arguments

```python
def multiply(x, y):
    return x * y

print(multiply(3, 4))  # 12
```

```python
def student_info(name, age):
    print(f"Name: {name}, Age: {age}")

student_info(age=20, name="Ravi")
```

```python
def greet(name="Guest"):
    print("Hello", name)

greet()          # Hello Guest
greet("Meena")   # Hello Meena
```

```python
def sum_numbers(x,y=10):    ## y=10 is a default parameter
    return x+y

print(sum_numbers(30, 40))
print(sum_numbers(50))
```

```python
def add_numbers(*nums):
    total = 0
    for n in nums:
        total += n
    return total

print(add_numbers(10, 20))
print(add_numbers(1, 2, 3, 4, 5))
```

```python
def student_info(**details):
    for key, value in details.items():
        print(f"{key}: {value}")

student_info(name="Meena", age=20, course="Python")
student_info(name="Radha", course="Java")
```

### 2. Scope of Variables

```python
def my_function():
    x = 10   # local variable
    print("Inside function, x =", x)

my_function()
# print(x)  # ❌ Error: x is local, not accessible outside
```

```python
x = 50   # global variable

def my_function():
    print("Inside function, x =", x)

my_function()
print("Outside function, x =", x)
```

```python
x = 100

def modify():
    global x   # declare x as global
    x = x + 50
    print("Inside function, x =", x)

modify()
print("Outside function, x =", x)
```

```python
x = 20

def my_function():
    x = 99   # local variable with same name
    print("Inside function, x =", x)

my_function()
print("Outside function, x =", x)
```

```python
def outer():
    x = 30
    
    def inner():
        nonlocal x
        x = 40
        print("Inside inner:", x)
    
    inner()
    print("Inside outer:", x)

outer()
```
