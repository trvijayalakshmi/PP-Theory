```python
# Create arithmetic.py file
def add(a, b):
    return a + b
def sub(a, b):
    return a - b
def mul(a, b):
    return a * b
def div(a, b):
    return a / b
```

```python
# Create geometry1.py file
def square(a):
    area_s = a*a
    peri_s = 4*a
    return area_s, peri_s
def rectangle(a,b):
    area_r = a*b
    peri_r = 2*(a+b)
    return area_r, peri_r
def triangle(a,b,c):
    area_t = 0.5 * a * b
    peri_t = a + b + c
    return area_t, peri_t
```
