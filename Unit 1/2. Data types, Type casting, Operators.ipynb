# 2. Numeric Data Types

There are three numeric types in Python:

int
float
complex
Variables of numeric types are created when you assign a value to them:

```python
x = 1    # int
y = 2.8  # float
z = 4+3j   # complex
```

```python
print(type(x))
print(type(y))
print(type(z))
```

a) Integers

```python
x = 1
y = 35656222554887711
z = -3255522

print(type(x))
print(type(y))
print(type(z))
```

b) Floating Point Numbers

```python
x = 35e3
y = 12E4
z = -87.7e-100

print(type(x))
print(type(y))
print(type(z))
print(y)
print(x)
print(z)
```

c) Complex numbers

```python
# Creating complex numbers
import numpy as np
z1 = 3 + 4j
z2 = 2 - 5j

print("z1 =", z1)
print("z2 =", z2)
print(np.real(z1))
```

# 3. Type Casting

    • Implicit Type Casting → Done automatically by Python.
    • Explicit Type Casting → Done manually by the programmer using built-in functions.

#### 1. Implicit Type Conversion

```python
x = 10     # int
y = 3.5    # float
z = x + y  # int + float → float
print(z)   # 13.5
print(type(z))  # <class 'float'>
```

#### 2. Explicit Type casting

```python
# Converting floating point to integer
x = int(3.14)
print(x) 
print(type(x))
```

```python
# Converting string to floating point
x = float("3.14")
print(x) 
print(type(x))
```

```python
# Converting integer to string
x = str(123)
print(x)
print(type(x))
```

```python
x = int("45")
print(type(x))
```

```python
# Converts a value to a boolean
x = bool(0)
print(x)
print(type(x))
```

# 4. Operators

1. Arithmetic Operators
2. Relational Operators
3. Logical Operators
4. Assignment Operators
5. Membership Operators
6. Identity Operators
7. Bitwise Operators

```python
# 1. Arithmetic Operators

x = -10
y = 4
result = x + y * 2 
print(result)

z1 = x //y
z2 = x /y
print(z1)
print(z2)
```

```python
# 2. Relational Operators

a, b = 10, 20
print("a > b :", a > b)     # False
print("a < b :", a < b)     # True
```

```python
x, y = 15, 15
print("x == y :", x == y)   # True
print("x != y :", x != y)   # False
```

```python
# 3. Logical Operators

x, y = 10, 0
print("x and y:", x and y)   # 0 (returns y since x is truthy)
print("x or y:", x or y)     # 10 (returns first truthy value)
```

```python
a, b = True, False
print("a and b:", a and b)   # False
print("a or b:", a or b)     # True
print("not a:", not a)       # False
```

```python
# Example combining all logical operators
x = 15

if (x > 10 and x < 20) and not (x == 15):
    print("Condition is True")
else:
    print("Condition is False")
```

```python
Augmented assignment
```

```python
# 4. Assignment operators

x = 20
x /= 5   # same as x = x / 5
print("Result", x)
```

```python
Multiple assignment
```

```python
# Assigning multiple variables in one line
a, b, c = 10, 20, 30
print("a =", a, "b =", b, "c =", c)   # a = 10 b = 20 c = 30

# Swapping two variables (no need of temp variable)
x, y = 5, 7
x, y = y, x
print("After swapping: x =", x, "y =", y)   # x = 7 y = 5

# Assigning same value to multiple variables
p = q = r = 50
print("p =", p, "q =", q, "r =", r)   # p = 50 q = 50 r = 50
```

```python
x = 17
x %= 5 
print("Result", x)
```

```python
# 5. Membership operators

x = [[1, 2], [3, 4]]
print(3 in x)  # Output: True
print(6 not in x)  # Output: True

y = [3, 4]
print(y in x)
```

```python
# 6. Identity operators

x = [1, 2, 3]
y = [1, 2, 3]

print(x == y)   # True → contents are the same
print(x is y)   # False → stored at different memory locations
print(x is not y)

z = 4
w = 4
print(z is w)
```

```python
# 7. Bit-wise operators

a = 12  # 1100
b = 6   # 0110

print("a & b =", a & b)   # 0100 → 4     (a and b)
print("a | b =", a | b)   # 1110 → 14     (a or b)
print("a ^ b =", a ^ b)   # 1010 → 10     (a xor b)
a = 20
print("a << 2 =", a << 3) # 110000 → 48   (Left shift)
print("a >> 2 =", a >> 2) # 0011 → 3    (Right shift)
```
