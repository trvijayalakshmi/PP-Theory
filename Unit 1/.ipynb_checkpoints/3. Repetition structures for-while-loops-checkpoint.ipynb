### Python Loops

Python has two primitive loop commands:

a) for loops
b) while loops

### a) The for loop

```python
for i in range(8):
    print(i)
```

```python
for i in range(2,8):
    print(i)
```

```python
for i in range(2,8,2):
    print(i)
```

```python
for i in range(7,2,-1):
    print(i)
```

```python
for i in range(1,10):
    if i == 4:
        break
    print(i)    
```

```python
for i in range(1,10):
    if i == 4:
        continue
    print(i) 
```

```python
for i in range(1,3):  # try changing 3 to 5
    if i == 3:
        break
    print(i)
else:
    print("end of for loop")
```

```python
for i in range(1,5):  
    if i == 3:
        continue
    print(i)
else:
    print("not in for loop")
```

```python
for i in range(1, 6):
    if i == 3:
        pass   # Do nothing when i = 3
    else:
        print(i)
```

```python
# Using pass statement in an empty loop
for i in range(5):
    pass   # Loop runs but does nothing
print("Loop finished!")
```

### b) The while Loop
With the while loop we can execute a set of statements as long as a condition is true.

###### The while loop requires relevant variables to be ready, in this example we need to define an indexing variable, i, which we set to 1.

```python
#Print i as long as i is less than 6:

i = 1
while i < 6:
    print(i)
    i += 1
```

```python
#Print a message once the condition is false:

i = 1
while i < 6:
    print(i)
    i += 1
else:
    print("i is no longer less than 6")
```

```python
# Exit the loop when i is 3:  Change 2 to 5

i = 0
while i < 2:
    i += 1
    if i == 3:
        break
    print(i)  
else:
    print("condition in while loop is false")  
```

```python
#Continue to the next iteration if i is 3:

i = 0
while i < 6:
    i += 1
    if i == 3:
        continue
    print(i)
else:
    print("not in while loop")
```

```python
i = 1
while i <= 3:
    pass
    i += 1

print("Loop finished!")
```

```python
# Right-angle triangle pattern
for i in range(1, 6):       # Outer loop controls rows
    for j in range(i):      # Inner loop prints stars in each row
        print("*", end=" ")
    print()
```

## Programs using for and while loops

```python
for i in range(5, 0, -1):  # Iterate from 5 to 1 in reverse order
    for j in range(i,6):    # Repeat the current number i times
        print(i, end=" ") # Print the current number followed by a space
    print() 
```

```python
for i in range(5, 0, -1):  
    for j in range(i):   
        print(i, end=" ") 
    print()
```

```python
s = 6
e = 30
for i in range(s, e):
        if i == 0 or i == 1:
            continue
        else:
            for j in range(2, int(i/2)+1):
                if i % j == 0:
                    break
            else:
                print(i,end = " ")
```

```python
n = int(input("Enter n: "))
a, b = 0, 1
print("Fibonacci Series:", end=" ")
for i in range(n):
    print(a, end=" ")
    a, b = b, a+b
```

```python
n = int(input("Enter n: "))
a, b = 0, 1
count = 0
while count < n:
    print(a, end=" ")
    a, b = b, a + b
    count += 1
```

```python
a = 64
b = 4
print(f"The gcd of {a} and {b} is", end=" ")
while b:
    a, b = b, a % b
print(a)
```

```python
total = 0
for i in range(1, 11):
    total += i
print("Sum of first 10 numbers is:", total)
```

```python
num = int(input("Enter a number: "))
for i in range(1, 11):
    print(f"{num} x {i} = {num*i}")
```

```python
num = int(input("Enter a number: "))
fact = 1
i = 1
while i <= num:
    fact *= i
    i += 1
print("Factorial is:", fact)
```

```python
num = int(input("Enter a number: "))
rev = 0
while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10
print("Reversed number is:", rev)
```

```python
text = input("Enter a string: ")
rev = ""
for ch in text:
    rev = ch + rev
print("Reversed String:", rev)
```

```python
# Prime numbers up to N using while and for loop
N = int(input("Enter a number N: "))

num = 2  # first prime number
while num <= N:   # while loop checks all numbers up to N
    is_prime = True
    for i in range(2, num):   # for loop checks divisibility
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num, end=" ")
    num += 1
```

```python
N = int(input("Enter N: "))

i = 1
print("Even numbers: ", end = " ")
while i <= N:
    if i % 2 == 0:
        print(i, end =" ")
    i += 1

i = 1
print("\nOdd numbers: ", end = " ")
while i <= N:
    if i % 2 != 0:
        print(i, end =" ")
    i += 1        
```

```python
N = int(input("Enter N: "))
even_sum, odd_sum = 0, 0
i = 1
while i <= N:
    if i % 2 == 0:
        even_sum += i
    else:
        odd_sum += i
    i += 1
print("Sum of Even numbers:", even_sum)
print("Sum of Odd numbers:", odd_sum)
```

```python
numbers = [3, 8, 15, 22, 7, 10]

for n in numbers:
    if n == 7:
        print("Number 7 found! Stopping search.")
        break
    print("Checking:", n)
```
