### Indentation
Python relies on indentation (whitespace at the beginning of a line) to define scope in the code. Other programming languages often use curly-brackets for this purpose.

```python
#If statement, without indentation (will raise an error):

a = 33
b = 200
if b > a:
print("b is greater than a") # you will get an error
```

### Elif
The elif keyword is pythons way of saying "if the previous conditions were not true, then try this condition".

```python
a = 33
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
else:
  print("b is less than a")  
```

# Else
The else keyword catches anything which isn't caught by the preceding conditions.

```python
a = 200
b = 330
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
else:
  print("a is greater than b")
```

```python
#You can also have an else without the elif:
a = 200
b = 33
if b > a:
  print("b is greater than a")
else:
  print("b is not greater than a")
```

### Short Hand If
If you have only one statement to execute, you can put it on the same line as the if statement.

```python
a = 200
b = 330
if a < b: print("a is greater than b")
```

### Short Hand If ... Else
If you have only one statement to execute, one for if, and one for else, you can put it all on the same line:

```python
a = 35
b = 33
print("A") if a > b else print("B")
```

```python
#One line if else statement, with 3 conditions:

a = 38
b = 38
print("A") if a > b else print("Equal") if a == b else print("B")
```

### And
The and keyword is a logical operator, and is used to combine conditional statements:

```python
#Test if a is greater than b, AND if c is greater than a:

a = 400
b = 330
c = 500
if a > b and c > a:
    print("Both conditions are True")
else:
    print("nothing")
```

### Or
The or keyword is a logical operator, and is used to combine conditional statements:

```python
#Test if a is greater than b, OR if a is greater than c:

a = 2000
b = 330
c = 5000
if a > b or a > c:
  print("At least one of the conditions is True")
else:
  print("Both conditions are false")  
```

### Nested If
You can have if statements inside if statements, this is called nested if statements.

```python
x = 21

if x > 10:
    print("Above ten")
    if x > 20:
        print("and also above 20!")
    else:
        print("but not above 20.")
```

```python
x = 8
if x > 0:
    if x % 2 == 0:
        print("Positive Even")
```

```python
# Example of if else statement with logical operators

# Variables
x = 5
y = 10
z = 5

# If...else statement with logical operators
if x == y and x == z:
    print("x is equal to both y and z")
elif x == y or x == z:
    print("x is equal to either y or z")
else:
    print("No condition matched")
```

```python
# Example of if else with boolean variables
is_weekend = True
is_sunny = False

# If...else statement with logical operators
if is_weekend or is_sunny:
    print("It's either the weekend or it's sunny. Time for outdoor activities!")
else:
    print("It's neither the weekend nor sunny. Maybe stay indoors.")
```

### The pass Statement
if statements cannot be empty, but if you for some reason have an if statement with no content, put in the pass statement to avoid getting an error.

```python
a = 10   # Try for a = 30
b = 20

if b > a:
    pass
else:
    print("a is less than or equal to b")  
```

```python
char = input("Enter a character: ")

if char.isdigit():
    print(f"{char} is a digit.")
elif char.islower():
    print(f"{char} is a lowercase character.")
elif char.isupper():
    print(f"{char} is an uppercase character.")
else:
    print(f"{char} is a special character.")
```

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

if num1 > num2:
    print(num1, "is the largest number.")
else:
    print(num2, "is the largest number.")
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a >= b and a >= c:
    largest = a
elif b >= a and b >= c:
    largest = b
else:
    largest = c

print("The largest number is:", largest)
```

```python
marks = int(input("Enter your marks (0-100): "))

if marks >= 90:
    grade = "A+"
elif marks >= 80:
    grade = "A"
elif marks >= 70:
    grade = "B"
elif marks >= 60:
    grade = "C"
elif marks >= 50:
    grade = "D"
elif marks >= 40:
    grade = "E"
else:
    grade = "F (Fail)"

print("Your grade is:", grade)
```

```python
year = int(input("Enter a year: "))

if year % 4 == 0:          # First check: divisible by 4
    if year % 100 == 0:    # If divisible by 100, then check further
        if year % 400 == 0:   # If divisible by 400 -> Leap Year
            print(year, "is a Leap Year")
        else:
            print(year, "is NOT a Leap Year")
    else:
        print(year, "is a Leap Year")
else:
    print(year, "is NOT a Leap Year")
```

```python
units = int(input("Enter electricity units consumed: "))

if units <= 100:
    bill = units * 5
elif units <= 200:
    bill = (100 * 5) + (units - 100) * 7
elif units <= 300:
    bill = (100 * 5) + (100 * 7) + (units - 200) * 10
else:
    bill = (100 * 5) + (100 * 7) + (100 * 10) + (units - 300) * 15

print("Total electricity bill is Rs.", bill)
```

```python
age = int(input("Enter your age: "))

if age < 5:
    price = 0
elif age <= 12:
    price = 50
elif age <= 18:
    price = 100
elif age <= 60:
    price = 200
else:
    price = 150

print("Ticket price: Rs.", price)
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Select operation: +, -, *, /, %")
choice = input("Enter your choice: ")

if choice == "+":
    result = num1 + num2
elif choice == "-":
    result = num1 - num2
elif choice == "*":
    result = num1 * num2
elif choice == "/":
    if num2 != 0:
        result = num1 / num2
    else:
        result = "Error! Division by zero."
elif choice == "%":
    result = num1 % num2
else:
    result = "Invalid choice"

print("Result:", result)
```
