### 1. Creating Dictionaries

```python
dict1 = {"name": "Alice", "age": 25, "city": "New York"}
dict2 = dict([("name", "Alice"), ("age", 25), ("city", "New York")])
dict3 = dict(name = "Alice", age = 25, city = "New York")
dict4 = {}  # empty dictionary

print(dict1)
print(dict2)
print(dict3)
print(dict4)
print(len(dict1))
print(len(dict4))
```

### 2. Accessing and Changing Dictionary Elements: Using keys.

```python
# Example 1
person = {"name": "John", "age": 30, "country": "India"}

print(person["name"])      # John
print(person.get("age"))
print(person.get("age",'Not found'))   # 30
```

```python
print(person.get("gender"))
print(person.get("gender", "Not Found"))  # Default value if key not found
```

```python
print(person.get('gender'))  # ✅ None
print(person['gender'])   # ❌ KeyError
```

```python
# Adding items to Dictionary
country_capitals = {
  "United States": "New York", 
  "Italy": "Naples" 
}

# add an item with "Germany" as key and "Berlin" as its value
country_capitals["Germany"] = "Berlin"
print(country_capitals)
```

```python
# Changing a value
country_capitals = {
  "United States": "New York", 
  "Italy": "Naples", 
  "England": "London"
}

# change the value of "Italy" key to "Rome"
country_capitals["Italy"] = "Rome"

print(country_capitals)
```

```python
# Change a single key
country_capitals = {
  "United States": "New York", 
  "Italy": "Naples", 
  "England": "London"
}

country_capitals['India'] = country_capitals.pop('Italy')
print(country_capitals)
```

### 3. Dictionary Methods

```python
student = {"id": 1, "name": "Ravi", "marks": 90}

print(student.keys())     # dict_keys(['id', 'name', 'marks'])
print(student.values())   # dict_values([1, 'Ravi', 90])
print(student.items())    # dict_items([('id', 1), ('name', 'Ravi'), ('marks', 90)])
```

```python
student.update({"marks": 95, "grade": "A"})
print(student)
```

```python
student.get("marks")
```

```python
student.pop("id")      # removes 'id'
print(student)
```

```python
student.popitem()      # removes last one
print(student)
```

```python
student.clear()        # clears dictionary
print(student)
```

```python
student = {"id": 1, "name": "Ravi", "marks": 90}

del student["name"]
print(student)
```

```python
del student
print(student)
```

```python
my_dict = {1: "Hello", "Hi": 25, "Howdy": 100}

print(1 in my_dict) 

# the not in operator checks whether key doesn't exist
print("Howdy" not in my_dict) 

print("Hello" in my_dict) 
# membership operator checks only keys and not values
```

### 4. Iterating Over Dictionaries

```python
person = {"name": "Rahul", "age": 22, "city": "Delhi"}
for key in person:
    print(key)
```

```python
for value in person.values():
    print(value)
```

```python
for key, value in person.items():
    print(key, ":", value)
```

### 5. Dictionary Comprehension

```python
squares = {x: x*x for x in range(1, 10)}
print(squares)   # {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
```

```python
squares = {x: x*x for x in range(1, 10) if x%2==0}
print(squares) 
```

```python
text = "banana"
# print(text.count('a'))
freq = {ch: text.count(ch) for ch in text}
print(freq)  
```

```python
text = "banana"

freq = {}
for ch in text:
    freq[ch] = freq.get(ch, 0) + 1

print("Character Count:", freq)
```

```python
labels = {n: ("Even" if n % 2 == 0 else "Odd") for n in range(1, 10)}
print(labels)
```

```python
original = {"a": 1, "b": 2, "c": 3}
inverted = {value: key for key, value in original.items()}
print(inverted)
```

```python
# Without Comprehension
my_dict = {"a": 1, "b": 2, "c": 3}
print("Original Dict", my_dict)

inverted_dict = {}
for key, value in my_dict.items():
    inverted_dict[value] = key
    
print("Inverted Dict", inverted_dict)
```

### 6. Programs on Dictionaries

```python
students = {}

# Input names and marks of 5 students
for i in range(3):
    name = input(f"Enter name of student {i+1}: ")
    marks = int(input(f"Enter marks of {name}: "))
    students[name] = marks

# Find the student with the highest marks
topper = max(students, key=students.get)

print("\nStudents and their marks:", students)
print("Topper:", topper, "with", students[topper], "marks")
```

```python
x = {"Alice": 85, "Bob": 92, "Charlie": 78}
# print(x.get("Bob"))

topper = max(x, key=x.get)
print("Topper is:", topper)
```

```python
keys = ["id", "name", "age"]
values = [101, "Ravi", 21]

student = dict(zip(keys, values))
print("Student Dictionary:", student)
```

```python
data = {"a": 104, "b": 205, "c": 104, "d": 130}

data2 = {}
for key, value in data.items():
    if value not in data2.values():
        data2[key] = value

print("After Removing Duplicates:", data2)
```

```python
text = "apple banana apple orange banana apple"
words = text.split()

freq = {}
for i in words:
    freq[i] = freq.get(i,0) + 1

print("Word Frequency:", freq)
```

```python
text = "programming"

freq = {}
for ch in text:
    freq[ch] = freq.get(ch, 0) + 1

print("Character Count:", freq)
```
