### 3. Set Operations

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# Union (A ∪ B)
print("Union:", A | B)        
print("Union method:", A.union(B))

# Intersection (A ∩ B)
print("Intersection:", A & B)  
print("Intersection method:", A.intersection(B))

# Difference (A - B)
print("Difference:", A - B)    
print("Difference method:", A.difference(B))

# Symmetric Difference (A Δ B → elements in A or B but not both)
print("Symmetric Difference:", A ^ B)  
```

### 5. Set Methods

```python
thisset = {"apple", "banana", "cherry"}
thisset.add("orange")
print(thisset)
```

```python
x = {"apple", "banana", "cherry"}
y = {"pineapple", "mango", "papaya"}

x.update(y)
print(x)
```

```python
#Add elements of a list to a set:

x = {"apple", "banana", "cherry"}
y = ["kiwi", "orange"]

x.update(y)
print(x)
```

```python
#Remove "banana" by using the remove() method:

x = {"apple", "banana", "cherry"}
x.remove("banana")
print(x)
```

```python
#Remove "banana" by using the discard() method:

x = {"apple", "banana", "cherry"}
x.discard("banana")
print(x)
```

```python
#Remove the last item by using the pop() method:

x = {"apple", "banana", "cherry"}
y = x.pop()

print(y)
print(x)
```

```python
#The clear() method empties the set:

x = {"apple", "banana", "cherry"}
x.clear()
print(x)
```

```python
x = {"apple", "banana", "cherry"}

del x
print(x)
```

## Programs on Sets

```python
numbers = [1, 2, 2, 3, 4, 4, 5, 6, 1, 7]
unique_numbers = set(numbers)   # removes duplicates
print("Unique elements:", unique_numbers)
```

```python
A = {"apple", "banana", "cherry"}
B = {"banana", "cherry", "mango"}

common = A & B   # Intersection
print("Common elements:", common)
```

```python
text = "programming"
unique_chars = "".join(set(text))
print("Unique characters:", unique_chars)
```

```python
sentence = "Python is easy and Python is powerful"
words = sentence.split()
unique_words = set(words)
print("Unique words:", unique_words)
print("Number of unique words:", len(unique_words))
```

```python
A = {1, 2, 3}
B = {4, 5, 6}

if A.isdisjoint(B):
    print("Sets are disjoint (no common elements)")
else:
    print("Sets are not disjoint")
```

```python
# Create two sets
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# Finding common elements
common = set1.intersection(set2)

# Finding elements present only in the first set
only_in_first = set1.difference(set2)

# Displaying results
print("Set 1:", set1)
print("Set 2:", set2)
print("Common Elements:", common)
print("Elements only in Set 1:", only_in_first)
```
