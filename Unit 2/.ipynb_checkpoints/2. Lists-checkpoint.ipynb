## 2. Properties of Lists

```python
my_list = [10, 20, 30, 40]
print(my_list)       # [10, 20, 30, 40]
print(my_list[0])    # 10 (first element)
```

```python
my_list = [1, 2, 3]
my_list[1] = 20      # Changing element at index 1
print(my_list)       # [1, 20, 3]
```

```python
my_list = [5, 10, 5, 20]
print(my_list)       # [5, 10, 5, 20]
```

```python
my_list = [10, "Hello", 3.14, True]
print(my_list) 
```

```python
my_list = [10, 20, 30, 40, 50]
print(my_list[2])       # 30 (indexing)
print(my_list[-1])      # 50 (negative index)
print(my_list[1:4])     # [20, 30, 40] (slicing)
```

```python
my_list = [1, 2, 3]
my_list.append(4)    # Adding element
print(my_list)       # [1, 2, 3, 4]
my_list.pop()        # Removing last element
print(my_list) 
```

## 4. List Comprehension 

```python
Syntax: [expression for item in iterable if condition]
```

### Programs on List Comprehension

```python
squares = []
for x in range(1,6):
    squares.append(x**2)
print(squares)
```

```python
squares = [x**2 for x in range(1, 6)]
print(squares)  
```

```python
evens = []
odds = []
for x in range(1,11):
    if x % 2 == 0:
        evens.append(x)
    else:
        odds.append(x)
print("The even nos are", evens)  
print("The odd nos are", odds)  
```

```python
evens = [x for x in range(1, 11) if x % 2 == 0]
print("The even nos are", evens)  
odds = [x for x in range(1, 11) if x % 2 != 0]
print("The odd nos are", odds)  
```

```python
[expression_if_true if condition else expression_if_false for item in iterable]
```

```python
nums = [1, 2, 3, 4, 5]

labels = ["Even" if x % 2 == 0 else "Odd" for x in nums]
print(labels)
```

```python
x = ["python", "is", "fun"]
upper_words = [i.upper() for i in x]
print(upper_words)
```

```python
x = "python is fun"
s = x.split(" ")
print(s)
upper_words = [i.upper() for i in s]
print(upper_words)
```

```python
x = "python is fun"
uppercase_chars = [i.upper() for i in x]
print("Uppercase chars:", uppercase_chars)
```

```python
nums = [x for x in range(1, 21) if x % 2 == 0 and x % 3 == 0]
print(nums)
```

```python
sentence = "List Comprehension Example"
vowels = [ch for ch in sentence if ch.lower() in "aeiou"]
print(vowels) 
```

```python
text = "abce123"

result = ["Vowel" if ch.lower() in "aeiou" else "Consonant"
          for ch in text if ch.isalpha()]

print(result)
```

```python
s = input("Enter a sentence: ")
words = s.split()
reversed_words = [word[::-1] for word in words]
print(" ".join(reversed_words))
```

```python
multiples_of_3 = [x for x in range(31) if x % 3 == 0]
print("Multiples of 3:", multiples_of_3)
```

```python
table = [[i*j for j in range(1, 6)] for i in range(1, 4)]
print("Multiplication table:", table)
```
