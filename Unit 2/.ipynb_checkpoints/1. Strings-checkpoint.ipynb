### 1. Creation and Accessing Strings

A string is a sequence of characters enclosed in single ('), double (") or triple quotes (''' / """).

```python
str1 = 'Hello'
str2 = "Python"
str3 = '''This is 
a multiline string'''
```

Indexing - Each character has an index, starting from 0.

```python
word = "Python"
print(word[0])   # P
print(word[3])   # h
```

```python
text = "Programming"
print(text[0:6])   # Progra
print(text[3:])    # gramming
print(text[:5])    # Progr
print(text[-3:])
print(text[-8:-4])
```

### 2. String Methods - Built-in methods:

1. Case Conversion

```python
s = "python programming"
print(s.upper())   # PYTHON PROGRAMMING
print(s.lower())   # python programming
print(s.title())   # Python Programming
print(s.capitalize()) # Python programming

print(len(s))         # 18
```

2. Checking Content

```python
s = "Python123"
print(s.isalpha())   # False (contains numbers)
print(s.isdigit())   # False (contains letters)
print(s.isalnum())   # True (letters + numbers)
print(" ".isspace()) # True
```

3. Searching and Replacing

```python
s = "I like Python"
print(s.find("like"))     # 2 (index where found)
print(s.replace("Python", "Java"))  # I love Java
```

4. Splitting and Joining

```python
s = "apple,banana,cherry"
fruits = s.split(",")    # ['apple', 'banana', 'cherry']
print(fruits)

joined = "-".join(fruits)
print(joined)  # apple-banana-cherry
```

### 3. String Formatting

1. Using f-strings:

```python
name = "Alice"
age = 22
print(f"My name is {name} and I am {age} years old.")
```

2. Using format():

```python
print("My name is {} and I am {} years old.".format(name, age))
```

### 4. String Iteration

```python
word = "PYTHON"
for ch in word:
    print(ch, end=" ")   # P Y T H O N
```

## Programs on Strings

1. Check if a String is Palindrome

```python
text = input("Enter a string: ")

if text == text[::-1]:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

2. Count Words in a String

```python
s = input("Enter a sentence: ")
words = s.split()
print("Number of words:", len(words))
```

3. Check if a String Contains Only Digits

```python
s = input("Enter a string: ")

if s.isdigit():
    print("String contains only digits")
else:
    print("String has non-digit characters")
```

4. Find the Longest Word in a Sentence

```python
s = input("Enter a sentence: ")
words = s.split()
longest = max(words, key=len)
print("Longest word:", longest)
```

5. Count Vowels and Consonants

```python
s = input("Enter a string: ").lower()
vowels = "aeiou"
v_count = 0
c_count = 0

for ch in s:
    if ch.isalpha():
        if ch in vowels:
            v_count += 1
        else:
            c_count += 1

print("Vowels:", v_count)
print("Consonants:", c_count)
```

6. Program to replace all spaces in a string with '-'

```python
text = input("Enter a string: ")

new_text = text.replace(" ", "-")

print("Modified string:", new_text)
```

7. Reverse Each Word in a Sentence

```python
s = input("Enter a sentence: ")
words = s.split()
reversed_words = [word[::-1] for word in words]
print(" ".join(reversed_words))
```

8. Program to count words in a string without using split()

```python
text = input("Enter a string: ")

count = 0
in_word = False

for char in text:
    if char != " " and not in_word:
        count += 1
        in_word = True
    elif char == " ":
        in_word = False

print("Number of words:", count)
```
