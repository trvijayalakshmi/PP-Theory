### 1. Creating a Tuple

```python
# Empty tuple
t1 = ()
print(t1)   # ()

# Tuple with integers
t2 = (1, 2, 3, 4)
print(t2)   # (1, 2, 3, 4)

# Tuple with mixed data types
t3 = (10, "apple", 3.14, True)
print(t3)   # (10, 'apple', 3.14, True)
```

```python
t5 = (10,)     # single element tuple
print(t5)      # (10,)

t6 = (10)      # just an integer
print(t6)      # 10
```

```python
nested = (1, (2, 3), (4, 5, 6))
print(nested)        
print(nested[1][1])  # (accessing inside nested tuple)
```

```python
list_data = [1, 2, 3, 4]
t7 = tuple(list_data)
print(t7)  

string_data = "Python"
t8 = tuple(string_data)
print(t8) 
```

```python
t = (10, 20, 30)
print(t[0])   # Output: 10 (first element)
print(t[2])   # Output: 30 (last element)
```

```python
t = (1, 2, 3)
# t[0] = 5   # ❌ Error: 'tuple' object does not support item assignment
print(t)      # Output: (1, 2, 3)
```

```python
t = (1, 2, 2, 3, 1)
print(t)  # Output: (1, 2, 2, 3, 1)
```

```python
t = (10, "Python", 3.14, [1, 2, 3])
print(t)  
```

### 2. Accessing Tuple Elements

```python
my_tuple = (10, 20, 30, 40, 50)

print(my_tuple[0])  
print(my_tuple[-1])   # Negative indexing
print(my_tuple[2]) 
```

```python
x = ("apple", "banana", "cherry")
for i in x:
    print(i)
```

```python
x = ("apple", "banana", "cherry")
for i in range(len(x)):
    print(x[i])
```

```python
fruits = ("apple", "banana", "cherry", "mango", "grape")

print(fruits[1:4])   # From index 1 to 3 → ('banana', 'cherry', 'mango')
print(fruits[:3])    # From start to index 2 → ('apple', 'banana', 'cherry')
print(fruits[2:])
```

### 3. Tuple Methods

```python
my_tuple = (10, 20, 30, 20, 40, 20)

print(my_tuple.count(20))   
print(my_tuple.index(30)) 
```

### 4. Operations with Tuples (not methods, but useful):

```python
numbers = (10, 5, 20, 15)

print(len(numbers))    # 4
print(min(numbers))    # 5
print(max(numbers))    # 20
print(sum(numbers))    # 50
print(sorted(numbers)) 
```

```python
fruits = ("apple", "banana", "cherry", "mango")

reversed_tuple = fruits[::-1]   # slicing with step -1
print("Reversed Tuple:", reversed_tuple)
```

```python
# We can use + sign to concatenate (just like strings).

tuple_1 = ('a', 'b')
tuple_2 = (3, 4)
print(tuple_1 + tuple_2)
```

```python
thistuple = ("apple", "banana", "cherry")
del thistuple
print(thistuple)
```

```python
# multiplying a tuple by an integer produces a tuple containing the 
# original tuple repeated that many times. 

a = (1, 7, 9, 8)
print(a * 2)
```

```python
#Create a new tuple with the value "orange", and add that tuple:
x = ("apple", "banana", "cherry")
y = ("orange",)
x += y

print(x)
```

```python
import sys
a_list = ['abc', 'xyz', 123, 231, 13.31, 0.1312]
a_tuple = ('abc', 'xyz', 123, 231, 13.31, 0.1312)
print('The list size:', sys.getsizeof(a_list), 'bytes')
print('The tuple size:', sys.getsizeof(a_tuple), 'bytes')
```

### 5. Programs using Tuples

```python
numbers = (10, 20, 30, 40, 50, 60, 70, 80)

print("Tuple:", numbers)
print("Elements at odd indices:")

# Accessing elements at odd indices using slicing
for i in range(1, len(numbers), 2):   # start from index 1, step = 2
    print(f"Index {i}: {numbers[i]}")
```

```python
# Creating a list
my_list = [5, 10, 15, 20, 25]

# Converting list to tuple
my_tuple = tuple(my_list)

# Printing tuple and its length
print("List:", my_list)
print("Tuple:", my_tuple)
print("Length of tuple:", len(my_tuple))
```

```python
x = ("apple", "banana", "cherry")
y = list(x)
y.append("orange")
x = tuple(y)
print(x)
```

```python
x = ("apple", "banana", "cherry")
print(x[1])
y = list(x)
y[1] = "kiwi"
x = tuple(y)
print(x)
```
